# Modelo de Lotka-Volterra para Depredador-Presa

El modelo de Lotka-Volterra es un sistema de ecuaciones diferenciales usado para describir la dinámica de las interacciones entre dos especies: una presa y un depredador. Este es el modelo básico:

$\textbf{Ecuaciones del Modelo}$

Sea $x(t)$ la población de presas en el tiempo $t$ y $y(t)$ la población de depredadores en el tiempo $t$. Las ecuaciones del modelo Lotka-Volterra son:

$ \begin{cases}
\frac{dx}{dt} = \alpha x - \beta xy \\
\frac{dy}{dt} = \delta xy - \gamma y
\end{cases}$

$\textbf{Parámetros}$

1. $\alpha$: Tasa de crecimiento de la población de presas en ausencia de depredadores.
2. $\beta$: Tasa a la cual los depredadores capturan a las presas.
3. $\gamma$: Tasa de mortalidad de los depredadores en ausencia de presas.
4. $\delta$: Tasa a la cual los depredadores incrementan en número al consumir presas.

$\textbf{Explicación}$

La ecuación $\frac{dx}{dt}=\alpha x - \beta xy$ describe el cambio en la población de presas. En ausencia de depredadores $(y=0)$, la población de presas crece exponencialmente a una tasa $\alpha$. La presencia de depredadores $(y \neq 0)$ reduce la tasa de crecimiento de las presas proporcionalmente al producto de las poblaciones de presas y depredadores.

La ecuación $\frac{dy}{dt} = \delta xy - \gamma y$ describe el cambio en la población de depredadores. En ausencia de presas $(x=0)$, la población de depredadores decae exponencialmente a una tasa $\gamma$. La presencia de presas $(x \neq 0)$ incrementa la tasa de crecimiento de los depredadores proporcionalmente al producto de las poblaciones de presas y depredadores.

$\textbf{¿Para qué se utiliza el modelo de Lotka-Volterra?}$

El modelo se utiliza para:
1. $\textit{Entender Dinámicas Ecológicas:}$ Proporciona una base teórica para entender cómo las poblaciones de depredadores y presas pueden influenciarse mutuamente y fluctuar con el tiempo.
2. $\textit{Investigación en Ecología y Biología:}$ Ayuda a los ecólogos y biólogos a estudiar y predecir comportamientos en ecosistemas naturales.
3. $\textit{Manejo de Recursos Naturales:}$ Informar estrategias de conservación y manejo de fauna, especialmente en áreas donde una especie depredadora y una presa están en interacción directa.

$\textbf{¿Quién desarrolló el Modelo de Lotka-Volterra?}$

El modelo fue desarrollado independientemente por dos científicos en la primera mitad del siglo XX:

1. $\textit{Alfred J. Lotka:}$ Un biofísico y químico estadounidense, quien introdujo las ecuaciones en 1925 en el contexto de la teoría de la auto-regulación biológica.
2. $\textit{Vito Volterra:}$ Un matemático italiano que formuló el modelo en 1926 mientras estudiaba las fluctuaciones en las poblaciones de peces en el Adriático durante la Primera Guerra Mundial.

$\textbf{Importancia y Limitaciones:}$


$\textit{Importancia:}$ El modelo es crucial en la ecología teórica porque fue uno de los primeros en describir matemáticamente la interacción biológica. Su simplicidad permite la comprensión básica de las dinámicas poblacionales.

$\textit{Limitaciones:}$ El modelo básico asume que el entorno es homogéneo y no cambia con el tiempo, que las tasas de interacción son constantes y que no hay competencia entre individuos de la misma especie, lo cual no es siempre es realista.

$\textbf{Aplicaciones:}$



1. $\textit{Ecología y Biología de la Conservación:}$ Usado para modelar y predecir dinámicas en reservas naturales y evaluar el impacto de introducir o eliminar especies.
2. $\textit{Estudios Epidemiológicos:}$ Adaptado para estudiar la propagación de enfermedades infecciosas, donde la inteeacción entre un patógeno (depredador) y una población susceptible (presa) sigue dinámicas similares.
3. $\textit{Economía y Sociología:}$ Aplicado metafóricamente para modelar la competencia entre empresas o la interacción entre grupos sociales.

$\textbf{Equilibirio de la población}$

El equilibrio de la población se producce en el modelo cuando ninguno de los niveles de población cambia, es decir, cuando ambas derivadas son iguales a 0:

$\begin{cases}
x(\alpha-\beta y)=0 \\
-y(\gamma-\delta x)=0
\end{cases}$

El sistema de ecuaciones anteriores nos da dos soluciones:


1. $\{x=0, y=0\}$
2. $\{x=\frac{\gamma}{\delta}, y=\frac{\alpha}{\beta}\}$

Por lo tanto, hay dos equilibrios en las ecuaciones.

La primera solución representa efectivamente la extinción de ambas especies. Si ambas poblaciones están en $0$, entonces seguirán estándolo indefinidamente. La segunda solución representa un punto fijo en el que ambas poblaciiones mantienen sus números actuales distintos de $0$ y, en el modelo simplificado, lo hacen indefinidamente. Los niveles de población en los que se logra este equilibrio dependen de los valores elegidos de los parámetros $\alpha$, $\beta$, $\gamma$ y $\delta$.

$\textbf{Retato Fase}$

El retrato fase del modelo de Lotka-Volterra muestra las trayectorias de las poblaciones de presas y depredadores en el plano  $(x, y)$ . Esto revela cómo las poblaciones evolucionan con el tiempo y si convergen o divergen hacia los puntos críticos.

1. Punto crítico $(0, 0)$: Las trayectorias se alejan del origen, indicando un crecimiento ilimitado de ambas poblaciones.
2. Punto crítico $(\frac{\gamma}{\delta}, \frac{\alpha}{\beta})$: Las trayectorias pueden ser espirales, indicando ciclos estables de crecimiento y disminución de ambas poblaciones, o divergir o converger hacia este punto, dependiendo de su estabilidad.

 

Gráfica de espacio fase para el problema de la presa del depredador para varias condiciones iniciales de la población de depredadores.

Supongamos que hay dos especies de animales, un babuino (presa) y un guepardo (depredador). Si las condiciones iniciales son 10 babuinos y 10 guepardos, se puede trazar la progresión de las dos especies a lo largo del tiempo; dados los parámetros de que las tasas de crecimiento y muerte de los babuinos son $1.1$ y $0.4$ mientras que las de los guepardos son $0.1$ y $0.4$ respectivamente. La elección del intervalo de tiempo es arbitraria.

También se pueden trazar soluciones paramétricamente como órbitas en el espacio de fase, sin representar el tiempo, pero con un eje que representa el número de presas y el otro eje que representa el número de depredadores para todos los tiempos.

Esto corresponde a eliminar el tiempo de las dos ecuaciones diferenciales anteriores para producir una sola ecuación diferencial

$\frac{dy}{dx}=-\frac{y}{x}\frac{\delta x-\gamma}{\beta y-\alpha}$

Relacionar las variables $x$ y $y$. Las soluciones de esta ecuación son curvas cerradas. Es susceptible de separación de variables: integrando

$\frac{\beta y-\alpha}{y}dy+\frac{\delta x-\gamma}{x}dx=0$

produce la relación implícita

$V=\delta x-\gamma \ln{x}+\beta y-\alpha\ln{y}$

donde $V$ es una cantidad constante que depende de las condiciones iniciales y se conserva en cada curva.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir símbolos y parámetros
x, y = sp.symbols('x y')
alpha, beta, gamma, delta = 0.3, 12, 34, 6

# Definir las ecuaciones del modelo de Lotka-Volterra
dx_dt = alpha * x - beta * x * y
dy_dt = delta * x * y - gamma * y

# Encontrar los puntos críticos
critical_points = sp.solve([dx_dt, dy_dt], (x, y))

# Matriz Jacobiana
jacobi_matrix = sp.Matrix([[dx_dt.diff(var) for var in (x, y)],
                           [dy_dt.diff(var) for var in (x, y)]])

# Función discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Función para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)
    
    if t > 0 and d > 0 and disc > 0:
        return "Región I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Región II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Región III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Región IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Región V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()
    
    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificación"

# Evaluación en puntos críticos y clasificación
print("Puntos críticos y su clasificación:")
for point in critical_points:
    J = jacobi_matrix.subs({x: point[0], y: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)
    
    print(f"\nPunto crítico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificación: {clas(J)}")

# Retrato fase del modelo de Lotka-Volterra
def lotka_volterra(X, t, alpha, beta, gamma, delta):
    x, y = X
    dxdt = alpha * x - beta * x * y
    dydt = delta * x * y - gamma * y
    return [dxdt, dydt]

x_range = np.linspace(0, 10, 20)
y_range = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x_range, y_range)
dX_dt = np.array([[dx_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)
dY_dt = np.array([[dy_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)

fig, ax = plt.subplots()
ax.streamplot(X, Y, dX_dt, dY_dt, density=1)
for point in critical_points:
    ax.plot(point[0], point[1], 'ro')
plt.xlabel('Presa (x)')
plt.ylabel('Depredador (y)')
plt.title('Retrato Fase del Modelo de Lotka-Volterra')
plt.grid()
plt.show()

# Graficar trayectorias del sistema de Lotka-Volterra
def grafica(dx_dt):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(1, 5, 5)
    for r in ic:
        for s in ic:
            x0 = [r, s]
            xs = odeint(lotka_volterra, x0, ts, args=(alpha, beta, gamma, delta))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Presa (x)', fontsize=15)
    plt.ylabel('Depredador (y)', fontsize=15)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 10)
    plt.ylim(0, 10)

    X, Y = np.mgrid[0:10:10j, 0:10:10j]
    u = alpha * X - beta * X * Y
    v = delta * X * Y - gamma * Y
    plt.quiver(X, Y, u, v, color='b')
    plt.show()

# Ejecutar la gráfica de trayectorias
grafica(lotka_volterra)
```

# Modelo Lotka-Volterra para Competencia Económica

Consideremos dos especies con tamaño de población $N_1(t)$ y $N_2(t)$, además asumimos que cada una crece logísticamente en ausencia de la otra. Cada especie tiene una tasa de crecimiento per cápita que ira decreciendo linealmente de acuerdo al tamaño de su población.

$\begin{cases}
\frac{dN_1}{dt}=N_1r_1(1-\frac{N_1}{K_1}) \\
\frac{dN_2}{dt}=N_2r_2(1-\frac{N_2}{K_2})
\end{cases}$ (1 y 2)


Además, los coeficientes $r_i$ y $K_i$ con $i=1,2$ representa la tasa de crecimiento pblacional de la especie y la capacidad de carga del ecosistema respectivamente. A estas condiciones le agregamos los factores de $\textit{competencia}$. Estos simplemente los parametrizamos con los coeficientes $\alpha_{12}$ y $\alpha{21}$ que describan la fuerza del efecto de la especie 2 sobre la especie 1 y de la especie 1 sobre la especie 2. Luego como esta competición afecta el tamaño de las poblaciones, sumamos el producto $\alpha_{ij}N_i$ al tamaño de su población en un instante $t$, esto es: 

$\begin{cases}
\frac{dN_1}{dt}=N_1r_1(1-\frac{N_1+\alpha_{12}N_2}{K_1}) \\
\frac{dN_2}{dt}=N_2r_2(1-\frac{N_2+\alpha_{21}N_1}{K_2})
\end{cases}$ (3)

En esta ecuación queda completamente caracterizado el modelo clásico de competencia de Lotka y Volterra para dos especies. Motivado por el biólogo Umberto D´anacona que quería modelar las dinámicas poblacionales en el mar adriático. Más tarde, este concepto es extendido más allá del depredador-presa y se obtiene este modelo generaizado de Lotka-Volterra, este tiene aplicaciones más alla de la biología y el estudio de las dinámicas poblacionales tales como las cienciads sociales, la economía e incluso las tecnologías.

$\textbf{Estabilidad Local}$

Como hemos visto, el estudio de estabilidad para el sistema (3) cerca de sus puntos de equilibrio equivale a hacerlo en la linealización del mismo, que resulta ser la matriz jacobiana del sistema. La cual presentaremos a continuación:

$J=\begin{bmatrix} \frac{r_1}{K_1}(K_1-N_1-\alpha_{12}N_2)-\frac{r_1}{K_1}N_1 & -\alpha_{12}\frac{r_1}{K_1}N_1 \\ -\alpha_{21}\frac{r_2}{K_2}N_2 & \frac{r_2}{K_2}(K_2-N_2-\alpha_{21}N_1)-\frac{r_2}{K_2}N_2 \end{bmatrix} $

En primer lugar hallaremos los puntos críticos $v=(x_0,y_0)$ del modelo (3), es decir cuando $N´_1=N´_2=0$.

Observemos que si $N_1=0$ ó $N_2=0$ podemos considerar entonces las ecuaciones de (1 y 2) de manera independiente respectivamente, y de la solución a la ecuación logística sabemos que estas se anulan cuando $N_i=K_i$. Procedemos al análisis de los casos en dónde se anula el sistema (3)

$\begin{cases} 
N_1r_1(1-\frac{N_1+\alpha_{12}N_2}{K_1})=0 \\
N_2r_2(1-\frac{N_2+\alpha_{12}N_1}{K_2})=0
\end{cases}$ (4 y 5)

1. Se puede notar que si $N_1=N_2=0$, las dos funciones se anulan simultáneamente, por lo tanto el sistema tiene un punto crítico en $(0,0)$. Este punto crítico se denomina $\textit{punto de extinción de las dos especies}$.
2. Si $N_1=0$, $N_2\neq0$, entonces se satisface la ecuación en (4) y de (5) obtenemos $N_2=K_2$. Luego el punto $(0, K_2)$ es un punto crítico del sistema. Dicho punto se conoce como $\textit{Exclusión Competitiva de}$ $N_1$.
3. Supongamos que $N_2=0$, $N_1\neq0$, entonces se satisface la ecuación en (5) y de (4) obtenemos $N_1=K_1$ entonces el punto $(K_1,0)$ es un punto crítico del sistema (3). Este punto se conoce como $\textit{Exclusión competitiva de}$ $N_2$.
5. Si $N_1\neq0$, $N_2\neq0$. Entonces se satisface (3) si y sólo si

   $\begin{cases}
   (1-\frac{N_1+\alpha_{12}N_2}{K_1})=0\\
   (1-\frac{N_2+\alpha_{21}N_1}{K_1})=0
   \end{cases}$

   Cuando $\alpha_{12}\alpha_{21}\neq 1$ la solución de este sistema lineal esta dada por

   $\begin{cases}
   N*_1=\frac{\alpha_{12}K_2 -K_1}{\alpha_{12}\alpha_{21}-1}\\
   N*_2=\frac{\alpha_{21}K_1 -K_2}{\alpha_{12}\alpha_{21}-1}
   \end{cases}$

   Entonces el sistema tiene un punto crítico en el punto $(N*_1,N*_2)$. Este caso será de especial interés para nosotros pues es el único bajo el cuál es posible la coexistencia de las dos

Por lo tanto, el sistema (3) cuenta con máximo cuatro puntos de equilibrio.

$\textbf{Suposiciones del Modelo}$

1. $\textit{Dos Empresas Competidoras:}$ Vamos a considerar dos empresas, $A$ y $B$, que compiten en el mismo mercado.
2. $\textit{Crecimiento Natural:}$ Cada empresa tiene una tasa de crecimiento natural en ausencia de competencia.
3. $\textit{Interacción Competitiva:}$ La competencia entre las empresas afecta negativamente a la tasa de crecimiento de cada una.

$\begin{cases}
\frac{dN_1}{dt}=N_1r_1(1-\frac{N_1+\alpha_{12}N_2}{K_1}) \\
\frac{dN_2}{dt}=N_2r_2(1-\frac{N_2+\alpha_{21}N_1}{K_2})
\end{cases}$ 

Donde:

1. $N_1$ es la cantidad de recursos (o cuota de mercado) que posee la empresa $A$
2. $N_2$ es la cantidad de recursos (o cuota de mercado) que posee la empresa $B$
3. $r_1$ y $r_2$ son las tasas de crecimiento natural de las empresas $A$ y $B$, respectivamente.
4. $K_1$ y $K_2$ son las capacidaders de carga del mercado para las empresas $A$ y $B$, respectivamente.
5. $\alpha_{12}$ y $\alpha_{21}$ son coeficientes que representan la intensidad de la competencia que la empresa $B$ impone sobre $A$ y viceversa.

$\textbf{Interpretación}$

1. Si $\alpha_{12}$ y $\alpha_{21}$ son grandes, la competencia es intensa y afecta significativamente las tasas de crecimiento de las empresas.
2. Si $\alpha_{12}$ y $\alpha_{21}$ son pequeños, la competencia es menos intensa y las empresas pueden crecer más independientemente.

Supongamos las siguientes variables:

1. $r_1=0.1$
2. $r_2=0.1$
3. $K_1=10$
4. $K_2=10$
5. $\alpha_{12}=0.1$
6. $\alpha_{21}=0.1$ 


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir símbolos y parámetros
x, y = sp.symbols('x y')
r_1, alpha_1, K_1, r_2, alpha_2, K_2 = 34, 45, 23, 2, 1, 10

# Definir las ecuaciones del modelo de Lotka-Volterra
dx_dt = r_1*x*(1-(x+alpha_1 * y)/K_1)
dy_dt = r_2*y*(1-(y+alpha_2 * x)/K_2) 

# Encontrar los puntos críticos
critical_points = sp.solve([dx_dt, dy_dt], (x, y))

# Matriz Jacobiana
jacobi_matrix = sp.Matrix([[dx_dt.diff(var) for var in (x, y)],
                           [dy_dt.diff(var) for var in (x, y)]])

# Función discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Función para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)
    
    if t > 0 and d > 0 and disc > 0:
        return "Región I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Región II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Región III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Región IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Región V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()
    
    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificación"

# Evaluación en puntos críticos y clasificación
print("Puntos críticos y su clasificación:")
for point in critical_points:
    J = jacobi_matrix.subs({x: point[0], y: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)
    
    print(f"\nPunto crítico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificación: {clas(J)}")

# Retrato fase del modelo de Lotka-Volterra
def lotka_volterra(X, t, r_1, alpha_1, K_1, r_2, alpha_2, K_2):
    x, y = X
    dxdt = r_1*x*(1-(x+alpha_1 * y)/K_1)
    dydt = r_2*y*(1-(y+alpha_2 * x)/K_2)
    return [dxdt, dydt]

x_range = np.linspace(0, 10, 20)
y_range = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x_range, y_range)
dX_dt = np.array([[dx_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)
dY_dt = np.array([[dy_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)

fig, ax = plt.subplots()
ax.streamplot(X, Y, dX_dt, dY_dt, density=1)
for point in critical_points:
    ax.plot(point[0], point[1], 'ro')
plt.xlabel('Empresa A')
plt.ylabel('Empresa B')
plt.title('Retrato Fase del Modelo de Lotka-Volterra para Competencia Económica')
plt.grid()
plt.show()

# Graficar trayectorias del sistema de Lotka-Volterra
def grafica(dx_dt, r_1, alpha_1, K_1, r_2, alpha_2, K_2):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(1, 5, 5)
    for r in ic:
        for s in ic:
            x0 = [r, s]
            xs = odeint(lotka_volterra, x0, ts, args=(r_1, alpha_1, K_1, r_2, alpha_2, K_2))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Empresa A', fontsize=15)
    plt.ylabel('Empresa B', fontsize=15)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 10)
    plt.ylim(0, 10)

    X, Y = np.mgrid[0:10:10j, 0:10:10j]
    u = r_1*X*(1-(X+alpha_1*Y)/K_1)
    v = r_2*Y*(1-(Y+alpha_2*X)/K_2)
    plt.quiver(X, Y, u, v, color='b')
    plt.show()

# Ejecutar la gráfica de trayectorias
grafica(lotka_volterra, r_1, alpha_1, K_1, r_2, alpha_2, K_2)
```

La gráfica muestra la evolución de las cuotas de mercado de las dos empresas, $A$ y $B$, a lo largo del tiempo según el modelo de competencia de Lotka-Volterra.

En este ejemplo, ambas empresas empiezan con la misma cuota de mercado y crecen en un entorno competitivo. La competencia afecta la tasa de crecimiento de cada empresa, reflejada en la interacción negativa entre $X$ e $Y$. En este caso particular, las cuotas de mercado fluctúan y pueden llegar a estabilizarse si las condiciones del mercado y la competencia se mantienen constantes.

Este modelo es una simplificación y no tiene en cuenta muchos factores reales, como cambios en las estrategias de mercado, innovaciones, fluctuaciones en la demanda, etc. Sin embargo, proporciona una base teórica para entender cómo la competencia puede influir en el crecimiento y la participación de mercado de las empresas.
