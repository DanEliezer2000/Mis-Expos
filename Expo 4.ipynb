# Modelo de Depredaci√≥n de Rosenzweigh y MacArthur

Es presentado en el art√≠culo _"Graphical Representation and Stability Conditions of Predator-Prey Interactions"_, donde, como su nombre lo indica, parten del an√°lisis de las representaciones gr√°ficas de la din√°mica depredador-presa, obteniendo como resultados principales la isoclina con pendiente cero de la presa, la cual tiene la forma parab√≥lica y los resultados condicionados a la respuesta funcional del depredador frente a la presa.

Este es un modelo que considera la densodependencia como factor y tambi√©n parten del modelo cl√°sico, pero en este caso no se considera una tasa de crecimiento constante de la presa, en lugar de eso se toma un crecimiento per c√°pita de la presa, y la tasa de muerte de las presas a causa de los depredadores es reemplazada por una respuesta funcional, ya que considera como inluye la cantidad de presas en el tiempo que emplea el depredador para cazar y para realizar otras actividades, obteniendo con esto el siguiente sistema:

$\begin{cases}
\frac{dx}{dt}=rx(1-\frac{x}{K})-yh(x) \quad\quad\quad\text{(4.1)} \\
\frac{dy}{dt}=y(-c+ph(x)) \quad\quad\quad\quad\quad\text{(4.2)}
\end{cases}$

Donde:

1. $x$ es la cantidad de presas
2. $y$ es la cantidad de depredadores
3. $r(1-\frac{x}{K})$ es la tasa de crecimiento per c√°pita de la poblaci√≥n de presas por unidad de tiempo
4. $K$ es la densidad m√°xima de la presa
5. $h(x)$ es la respuesta funcional que representa a la presa capturada por depredador por unidad de tiempo
6. $p$ es la tasa de producci√≥n de depredadores por presa consumida por unidad de tiempo
7. $c$ es la tasa de mortalidad de los depredadores en ausencia de las presas por unidad de tiempo.

Hay que observar que $r, K, c, d>0$, pues de no ser as√≠ el modelos carecer√≠a de sentido.

__Importancia__

Este modelo es de gran importancia ya que muestra a la densodependencia como un factor que influye en la interacci√≥n, adem√°s de mostrar que ni la tasa de nacimiento ni la de muerte son constantes, sino que dependen de la densidad poblacional que existe tanto de presas como de depredadores.

__An√°lisis del Sitema__

$\begin{cases}
\frac{dx}{dt}=rx(1-\frac{x}{K})-yh(x) \quad\quad\quad\text{(4.1)} \\
\frac{dy}{dt}=y(-c+ph(x)) \quad\quad\quad\quad\quad\text{(4.2)}
\end{cases}$

Donde:

1. $x$ es la cantidad de presas
2. $y$ es la cantidad de depredadores
3. $r(1-\frac{x}{K}$ es la tasa de crecimiento per c√°pita de la poblaci√≥n de presas por unidad de tiempo
4. $K$ es la densidad m√°xima de la presa
5. $h(x)$ es la respuesta funcional que representa a la presa capturada por depredador por unidad de tiempo
6. $p$ es la tasa de producci√≥n de depredadores por presa consumida por unidad de tiempo
7. $c$ es la tasa de mortalidad de los depredadores en ausencia de las presas por unidad de tiempo.

Este modelo considera que despu√©s de un tiempo $T$ lo suficientemente grande, un depredador mata un promedio de $N$ presas de una poblaci√≥n de $x$, por lo tanto $h(x)=\frac{N}{T}$, que como podemos ver se trata de una respuesta funcional de tiempo, pues cuando la cantidad de presas aumenta el depredador tarda menos en buscarlas, por lo cual hay que considerar tambi√©n el tiempo que el depredador no se encuentra cazando, $\tau$, adem√°s hay que notar que al igual que el tiempo, el espacio que recorre para encontrar una presa tambi√©n disminuye o aumenta dependiendo de la cantidad de presas. Llamemos $s$ a la tasa de las uniddades de √°rea de busqueda de presas, por unidad de tiempo, es decir, el √°rea que recorre en busca de alimento durante una unidad de tiempo.

Partiendo de estas observaciones, las cuales fueron realizadas por Smith, consideremos lo siguiente:

El tiempo que utiliza el depredador en buscar presar es $T-N\tau$, esto lo hace en un √°rea de $s(T-N\tau)$, lo que lo lleva a que los encuentros con la presa sean $xs(T-N\tau)$, pues este depende de la cantidad de presas disponibles, de donde se obtiene que la cantidad de presas que mata un depredador en cierto tiwmpo es

$N=xs(T-N\tau)$

por lo que

$T=\frac{N}{T}=\frac{N}{\frac{N(1-xs\tau)}{xs}}=\frac{xs}{1-xs\tau}$

Al sustituir esto en las ecuaciones origiles tenemos:

$\begin{cases}
\frac{dx}{dt}=rs(1-\frac{x}{K})-\frac{sxy}{1-xs\tau} \quad\quad\quad\text{3} \\
\frac{dy}{dt}=-cy+p(\frac{sxy}{1-xs\tau} \quad\quad\quad\quad\text{4}
\end{cases}$

Despu√©s de realizar los cambios de variables

1. $u=\frac{x}{X}$
2. $v=\frac{y}{Y}$

y seleccionar $X, Y$, tales que:

1. $s\tau X=1$
2. $Y=dX$
3. $m=\frac{d}{r}$
4. $k=\frac{K}{X}$

A partir de 3 obtenemos el siguiente sistema:

$\begin{cases}
\frac{dx}{dt}=x(1-\frac{x}{k})-\frac{mxy}{1+x} \quad\quad\quad\text{5} \\
\frac{dy}{dt}=-cy+\frac{mxy}{1+x} \quad\quad\quad\quad\text{6}
\end{cases}$

__Puntos de equilibrio__

En este sistema los puntos de equilibrio est√°n dados por las ecuaciones 3 y 4 igualadas a cero, es decir:

$\begin{cases}
x(1-\frac{x}{k}-\frac{my}{1+x})=0 \quad\quad\quad\text{7} \\
y(-c+\frac{mx}{1+x})=0 \quad\quad\quad\quad\text{8}
\end{cases}$

Esto indica que los puntos de equilibrio son el punto trivial, $(0,0)$ y el punto semitrivial, $(k,0)$

Pero observemos qu√© ocurre en estos puntos, en el caso trivial ambas poblaciones se extinguen debido a que este punto indica que las dos especies se quedaron sin representantes, mientas que en el caso semitrivial la poblaci√≥n del depredador es inexistente y como $K$ es la capicidad de carga de la presa, por lo que ante la ausencia de depredadores la poblaci√≥n de la presa crece siguiendo el modelo de Verhulst, por lo que al alcanzar e inclusive rebasar su l√≠mite ocasiona que el h√°bitat ya no sea capaz de mantenerla originando que tambi√©n se extinga, a√∫n ante la ausencia de depredador-presa, podemos observar que en estos puntos ni siquiera se da una intenci√≥n de este tipo, pues en el punto trivial no existe poblaci√≥n alguna y en e√± semitrivial s√≥lo se encuentra la presa, y ante la ausencia del depredador no se puede dar dicha interacci√≥n, por lo que es necesario verificar si hay puntos de coexistencia de equilibrio, es decir, donde tanto la poblaci√≥n de la presa como la del depredador tengan representantes, garantizando as√≠u que se dar√° una interacci√≥n como tal.

Finalmente, el punto de equilibrio no trivial es el que viene dado por:

$(x*,y*)=(\frac{c}{m-c},\frac{1+\frac{c}{m-c}-\frac{\frac{c}{m-c}}{k}-(\frac{\frac{c}{m-c}})^2{k}}{m})$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir s√≠mbolos y par√°metros
x, y = sp.symbols('x y')
k, m, c = 10, 0.5, 0.4  # Par√°metros del modelo

# Definir las ecuaciones del modelo de Rosenzweig-MacArthur
dx_dt = x * (1 - x / k) - (m * x * y) / (1 + x)
dy_dt = -c * y + (m * x * y) / (1 + x)

# Encontrar los puntos cr√≠ticos
critical_points = sp.solve([dx_dt, dy_dt], (x, y))

# Matriz Jacobiana
jacobi_matrix = sp.Matrix([[dx_dt.diff(var) for var in (x, y)],
                           [dy_dt.diff(var) for var in (x, y)]])

# Funci√≥n discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Funci√≥n para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)
    
    if t > 0 and d > 0 and disc > 0:
        return "Regi√≥n I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Regi√≥n II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Regi√≥n III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Regi√≥n IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Regi√≥n V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()
    
    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificaci√≥n"

# Evaluaci√≥n en puntos cr√≠ticos y clasificaci√≥n
print("Puntos cr√≠ticos y su clasificaci√≥n:")
for point in critical_points:
    J = jacobi_matrix.subs({x: point[0], y: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)
    
    print(f"\nPunto cr√≠tico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificaci√≥n: {clas(J)}")

# Retrato fase del modelo de Rosenzweig-MacArthur
def rosenzweig_macarthur(X, t, k, m, c):
    x, y = X
    dxdt = x * (1 - x / k) - (m * x * y) / (1 + x)
    dydt = -c * y + (m * x * y) / (1 + x)
    return [dxdt, dydt]

x_range = np.linspace(0, 10, 20)
y_range = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x_range, y_range)
dX_dt = np.array([[dx_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)
dY_dt = np.array([[dy_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)

fig, ax = plt.subplots()
ax.streamplot(X, Y, dX_dt, dY_dt, density=1)
for point in critical_points:
    ax.plot(point[0], point[1], 'ro')
plt.xlabel('Presa (x)')
plt.ylabel('Depredador (y)')
plt.title('Retrato Fase del Modelo de Rosenzweig-MacArthur')
plt.grid()
plt.show()

# Graficar trayectorias del sistema de Rosenzweig-MacArthur
def grafica(rosenzweig_macarthur):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(1, 5, 5)
    for r in ic:
        for s in ic:
            x0 = [r, s]
            xs = odeint(rosenzweig_macarthur, x0, ts, args=(k, m, c))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Presa (x)', fontsize=15)
    plt.ylabel('Depredador (y)', fontsize=15)
    plt.tick_params(labelsize=10)
    plt.xlim(2, 6)
    plt.ylim(4, 8)

    X, Y = np.mgrid[2:6:10j, 4:8:10j]
    u = np.array([[dx_dt.subs({x: X[i, j], y: Y[i, j]}) for j in range(10)] for i in range(10)], dtype=float)
    v = np.array([[dy_dt.subs({x: X[i, j], y: Y[i, j]}) for j in range(10)] for i in range(10)], dtype=float)
    plt.quiver(X, Y, u, v, color='b')
    plt.show()

# Ejecutar la gr√°fica de trayectorias
grafica(rosenzweig_macarthur)
```

__Conclusiones__

El motivo para crear nuevos modelos no es porque los anteriores no funcionen, si no que el fen√≥meno que se desea estudiar necesita datos m√°s precisos y los modelos existentes no logran obtenerlos, por lo que es necesario crear un modelo que se adapte mejor a la situaci√≥n, ya que siempre hay que tener en cuenta que el modelo no ser√° igual a la realidad, pues esto resultar√≠a imposible, debido a la cantidad de par√°metros que se tendr√≠an que consideran, por lo que al elaborar un modelo hay que tener claro que es lo que se desea saber, para establecer que par√°metros ser√°n considerados, y as√≠ poder obtener los datos deseados.

En el caso del modelo de Rosenzweig-MacArthur se consideran, especies con rol fijo, es decir, que √∫nicamente es presa o depredador, lo que lo hace diferente del modelo de intercambio de roles, adem√°s de tomar en cuenta la capacidad de carga, lo que lo hace un modelo con densodependencia, diferenci√°ndolo del modelo de Lotka-Volterra, por otro lado mientras que el modelo de Leslie-Gower presenta una respuesta funcional tipo I, Rosenzweig y MacArthur emplean un respuesta funcional de tipo II, ya que consideran que el depredador que utiliza menos tiempo para cazar, cuando hay m√°s presas. En este caso se tomaron como base los modelos de Lotka-Volterra, Verhulst y el modelo cl√°sico depredador-presa con densodependencia, para establecer el comportamiento que iba a surgir en la interacci√≥n al modificar la relaci√≥n que existe entre¬†los¬†par√°metros.

Al realizar un an√°lisis del sistema, se encontr√≥ que inicialmente este sistema s√≥lo ten√≠a como puntos de equilibrio al trivial, $(0,0)$ y al semi-trivial $(k, 0)$, sin embargo tiene un punto de coexistencia de equilibrio, $(x*,y*)$, que es aquel en donde se cruzan las isoclinas de grado cero, y dependiendo de las condiciones del par√°metro k y su relaci√≥n con los otros par√°metros, que influyen en el punto $(x*, y*)$, es lo que define el comportamiento de la interacci√≥n arrojando nuevamente tres casos: un sumidero, un punto fuente y centros, que es lo mismo que establece, Rosenzweig y MacArthur en su art√≠culo.

Si bien el trabajo trata sobre como influya la variaci√≥n de los par√°metros en el comportamiento del modelo, hay que notar que no basta con la modificaci√≥n de un par√°metro, para que el comportamiento cambie totalmente, sino que hay que cambiar la relaci√≥n que existe entre los par√°metros, pues es √©sta la que realmente influye para que se de un cambio considerable en comportamiento del sistema, ya que si solo se cambia el valor de un par√°metro, sin afectar a la relaci√≥n que guarda con los otros, el comportamiento no se altera significativamente.

Al recordar lo expuesto sobre densodependencia, se dijo que las poblaciones tienden a regresar a su estabilidad y podemos ver al ciclo l√≠mite como esta esta-bilidad, pues hablamos de dos poblaciones, no s√≥lo de una que fue para la que definimos estabilidad, donde se trataba de la capacidad de carga, y justo al considerar la capacidad de carga, al volver a considerar a las especies por separado, si los depredadores sobrepasan la capacidad de carga,logrando eliminar a la pre-sa, al quedarse sin alimento ellos tambi√©n perecer√°n, es decir, llegar√°n al punto de equilibrio¬†trivial.
